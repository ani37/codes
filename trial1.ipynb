```python
import pandas as pd
import numpy as np
import nltk
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
import sys
```

```python
df = pd.read_csv('testdata.manual.2009.06.14.csv', header=None,parse_dates=[2])
print(df.head)
```

```python
df_cols=['polarity', 'id','date','query','user','text']
df.columns = df_cols
df.columns
```

```python
df.drop(['polarity', 'id','date','query','user'], axis=1, inplace=True)
df.head()
```

```python
import string
punctuations = list(string.punctuation)
punctuations.append('...')
punctuations.append('.')
punctuations.append('..')
punctuations
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.tokenize import TweetTokenizer#specially tokenize twitter tweets
#from nltk.tokenize import RegexpTokenizer
```

```python
#tokenizing--word and sentence tokenizer(seperates by word and sentences respectively)
#corpora--body of text eg:medical journal,anything in english language
#lexicon--word and their meaning
#investor-speak and regular-english-speak
df.head()
```

```python
#tokenise
tt=TweetTokenizer()
df['text']=df['text'].apply(tt.tokenize)
```

```python
df['text-filtered'] = ""
```

```python
#remove stop words
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
stop_words.add('I')
stop_words.add('2')
for i in df.index:
    filtered_sentence = [w for w in df['text'][i] if not w in stop_words] 
    df['text-filtered'][i] = filtered_sentence
```

```python
stop_words
```

```python
#remove punctuations
for w in df.index:
    rem=[i for i in df['text-filtered'][w] if i not in punctuations]
    df['text-filtered'][w]=rem
```

```python
print(df['text'][0])
print(df['text-filtered'][0])
```

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
porter_stemmer = PorterStemmer()
df['text-stem']=df['text-filtered'].apply(lambda x : [porter_stemmer.stem(y) for y in x])
```
